## 함수 사용하기

- 함수 이름: mycalc
- 기능: 입력된 배열 값 모두 더한 후 리턴하기

```
def 함수이름(파라메터):
    return(리턴값)
```

````python
x = range(10)
print(x)
print(type(x))
````

````python
x = [0,1,2,3,4]
mycalc(x) 
# 
````

````python
def mycalc(x):
    sum = 0 
    for i in x:
        sum = sum + i
    return(sum)
````

````python
x = range(10)
mycalc(x)
````
